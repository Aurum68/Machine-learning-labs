1. Все элементы, большие некоторого числа `k` (вводится с клавиатуры).

```python
n = int(input("Input number of rows: "))
m = int(input("Input number of columns: "))

arr = np.random.random(size=(n, m)) * 100
arr
```

```python
print('Общее мат. ожидание: ', arr.mean())
print('Мат. ожидание по столбцам: ', arr.mean(axis=0))
print('Мат. ожидание по строкам: ', arr.mean(axis=1))
print('Общая дисперсия: ', arr.var())
print('Дисперсия по столбцам: ', arr.var(axis=0))
print('Дисперсия по строкам: ', arr.var(axis=1))
print('Общее стандартное отклонение: ', arr.std())
print('Стандартное отклонение по столбцам: ', arr.std(axis=0))
print('Стандартное отклонение по строкам: ', arr.std(axis=1))
```

**Задание 3** Даны два вектора одинакового размера. Найдите:

*  Косинус угла между векторами.
*  Евклидово расстояние между векторами.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

cosinus = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
cosinus
```

```python
euclidean = np.linalg.norm(a - b)
euclidean
```

**Задание 7** Загрузите файл `tips.csv`. Рассчитайте все одномерные статистические характеристики по столбцу `tip`.

**Указание:** см.  файл `4 Math.ipynb`

```python
tips = pd.read_csv("lab1/tips.csv", sep =',')
tips.head(10)
```

```python
print('Математическое ожидание', tips.tip.mean())
print('Дисперсия', tips.tip.var())
print('Ст. отклонение', tips.tip.std())
print('Минимум', tips.tip.min())
print('Максимум', tips.tip.max())
print('Сумма', tips.tip.sum())
print('Медиана', tips.tip.median())
print('Мода (самое частое значение)', tips.tip.mode())
print('Счетчик', tips.tip.count())
print('Квантиль (25%)', tips.tip.quantile(0.25))
```
